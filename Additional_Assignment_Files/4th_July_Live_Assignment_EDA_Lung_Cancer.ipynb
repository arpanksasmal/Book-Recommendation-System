### Age Distribution Analysis

### Gender Distribution Analysis

```python
df.GENDER.value_counts()
```

The dataset has a nearly equal number of males (1,512) and females (1,486). The gender distribution is balanced with only a slight difference between the two.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="GENDER", data=df, hue='GENDER')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Lung Cancer Distribution by Gender

```python
gender_lung_cancer_counts = df.groupby(['GENDER', 'LUNG_CANCER']).size().unstack().fillna(0)
gender_lung_cancer_counts
```

The number of females with lung cancer (764) is slightly higher than males (753), while more males (759) than females (722) do not have lung cancer.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="GENDER", hue="LUNG_CANCER", data=df)
plt.title('Lung Cancer Distribution Across Genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Lung Cancer', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

The count plot shows the distribution of lung cancer status by gender. According to the data:

There are 722 females and 764 females with lung cancer. There are 759 males and 753 males with lung cancer.

This indicates that the number of females with lung cancer is slightly higher than the number of males, while the number of males and females without lung cancer is relatively close.

This visualization provides an overview of how lung cancer prevalence varies between genders in the dataset.

### Countplots of Symptoms Associated with Lung Cancer

```python
df.columns
```

```python
X = ['YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 
     'WHEEZING', 'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH', 'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']

fig, ax = plt.subplots(nrows=4, ncols=3)
fig.set_size_inches(16, 24)

for i in range(4):
    for j in range(3):
        index = 3 * i + j
        if index < len(X):
            sns.countplot(x=df[X[index]], hue=df['LUNG_CANCER'], ax=ax[i][j])
            ax[i][j].set_title(f'Countplot of {X[index]}')

plt.tight_layout()
plt.show()
```

After analyzing the count plots, it’s evident that symptoms like yellow fingers, coughing, fatigue, shortness of breath, and chest pain are prevalent among respondents, indicating that these symptoms could be key indicators of lung cancer risk.

## Additional Visualizations: Box Plot, Violin Plot, Histogram, Pair Plot, Facet Grid, and KDE Plot

### Box Plot of Age Distribution by Lung Cancer Status

```python
age_stats_by_lung_cancer = df.groupby('LUNG_CANCER')['AGE'].describe()
age_stats_by_lung_cancer
```

The age distribution for people with and without lung cancer is similar, with mean ages of approximately 55.7 years for non-cancer and 54.7 years for cancer patients; both groups have ages ranging from 30 to 80.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Box Plot of Age by Lung Cancer Status')
plt.show()
```

The countplot shows that the age distribution for people with and without lung cancer overlaps significantly, with both groups spanning similar age ranges.

### Violin Plot of Age Distribution by Lung Cancer Status

```python
age_lung_cancer_summary = df.groupby('LUNG_CANCER')['AGE'].describe()
age_lung_cancer_summary
```

The average age of individuals with and without lung cancer is similar, with the mean age being slightly higher for those without lung cancer. Both groups have a similar range of ages, with most individuals falling between 30 and 80 years.

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Violin Plot of Age by Lung Cancer Status')
plt.show()
```

The countplot shows similar age distributions for individuals with and without lung cancer, with no significant differences in age ranges between the two groups.

### Histogram of Age Distribution by Lung Cancer Status

```python
age_histogram = df.groupby('LUNG_CANCER')['AGE'].describe()
age_histogram
```

The average age for individuals with lung cancer (54.7 years) is slightly lower than for those without lung cancer (55.7 years), with similar age ranges and distributions observed in both groups.

```python
plt.figure(figsize=(12, 8))
for category in df['LUNG_CANCER'].unique():
    subset = df[df['LUNG_CANCER'] == category]
    sns.histplot(subset['AGE'], kde=True, label=category)
plt.title('Histogram of Age by Lung Cancer Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Lung Cancer')
plt.show()
```

The age distribution for both lung cancer and non-lung cancer groups shows that most individuals are between 30 and 80 years old, with overlapping age ranges and no significant age difference between the two groups.

### Pair Plot of Selected Numerical Features

```python
numerical_features = ['AGE', 'FATIGUE', 'COUGHING', 'SHORTNESS_OF_BREATH']
selected_df = df[numerical_features + ['LUNG_CANCER']]

grouped_statistics = selected_df.groupby('LUNG_CANCER').describe().T
grouped_statistics
```

For all numerical features, the distributions are similar between patients with and without lung cancer, though there are minor differences in means and percentiles for AGE, COUGHING, and FATIGUE.

```python
sns.pairplot(df, hue='LUNG_CANCER', vars=['AGE', 'FATIGUE', 'COUGHING', 'SHORTNESS_OF_BREATH'])
plt.title('Pair Plot of Features by Lung Cancer Status')
plt.show()
```

The pair plot shows that while AGE, COUGHING, FATIGUE, and SHORTNESS_OF_BREATH have similar distributions, patterns and correlations among features might still differ by lung cancer status.

### Facet Grid of Selected Numerical Features by Lung Cancer Status

```python
numerical_features = ['AGE', 'FATIGUE', 'COUGHING', 'SHORTNESS_OF_BREATH']
summary_stats = df[numerical_features].describe()
summary_stats
```

 Age, Fatigue, Coughing, and Shortness of Breath show limited variability, with most values clustering around 1 and 2 for symptoms.

```python
g = sns.FacetGrid(df, col='LUNG_CANCER', height=5, aspect=1.2)
g.map(sns.histplot, 'AGE', kde=True)
g.set_axis_labels('Age', 'Frequency')
g.set_titles('Distribution of Age by Lung Cancer Status')
plt.show()
```

he facet grid displays the distribution of age separately for each lung cancer status, revealing age distribution differences between those with and without lung cancer.

### KDE Plot of Age Distribution by Lung Cancer Status

```python
age_no = df[df['LUNG_CANCER'] == 'NO']['AGE']
age_yes = df[df['LUNG_CANCER'] == 'YES']['AGE']
age_no.describe(), age_yes.describe()
```

The average age is slightly lower for those with lung cancer (54.65 years) compared to those without (55.70 years), with both groups having a similar age range.

```python
plt.figure(figsize=(12, 8))
for category in df['LUNG_CANCER'].unique():
    subset = df[df['LUNG_CANCER'] == category]
    sns.kdeplot(subset['AGE'], label=category, fill=True)
plt.title('KDE Plot of Age by Lung Cancer Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Lung Cancer')
plt.show()
```

The KDE plot shows overlapping distributions for age, with a slight shift indicating a lower average age for individuals with lung cancer.

### Correlation Matrix of Numerical Features

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numerical_df.corr()
correlation_matrix
```

The correlation matrix reveals that most features have weak correlations with each other. For example, AGE shows a small negative correlation with SMOKING and ALCOHOL_CONSUMING. FATIGUE has a small positive correlation with ALLERGY and CHEST_PAIN. This suggests that while there are some small relationships between features, no strong correlations are evident.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Having seen the heatmap, Not much correlation between different features but we can figure out something like:

1.Smoking leads to chronic desease

2.Alcohol consumption may lead to yellowish Fingers.

3.Alcohol may lead to Swallowing Difficulties

4.Fatigue and Anxiety may lead to Shortness of breath.
